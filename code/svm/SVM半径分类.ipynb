```python
import numpy  as np
from  sklearn.svm import  SVC #分类
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train=np.random.randn(300,2) #300个点 ，n正负，
x_train
```

```python
y_train=np.logical_xor(x_train[:,0]>0,x_train[:,1]>0)
y_train
```

```python
a=np.array([[1,2],[1,-2],[-1,-2],[-1,2]])
b=np.logical_xor(a[:,0]>0,a[:,1]>0)
b
```

```python
x=np.arange(-3,3,0.01) #600个点
y=np.arange(-3,3,0.01)  #600个点
xx,yy=np.meshgrid(x,y)#生成网格
xy=np.c_[xx.ravel(),yy.ravel()] #串联
xy.shape
```

```python
display(x_train.shape,y_train.shape)
```

```python
svc=SVC(kernel="rbf")#半径计算
svc.fit(x_train,y_train) #训练数据
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
dd=svc.decision_function(xy)
d=dd.reshape(xx.shape)
plt.imshow(d)
plt.contour(xx,yy,d)
```

```python
plt.figure(figsize=(8,8))

# 样本X到分离超平面的距离
Z = svc.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.imshow(Z,extent=(xx.min(), xx.max(), yy.min(), yy.max()), 
            cmap=plt.cm.PuOr_r)
#绘制轮廓
contours = plt.contour(xx, yy, Z)

plt.scatter(x_train[:,0], x_train[:,1], s=30, c=y_train, cmap=plt.cm.Paired,  edgecolors='k')
plt.axis('off')
plt.axis([-3, 3, -3, 3])
plt.show()

```
