```python
import sklearn.datasets as datasets
from  sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import  Axes3D #绘制3D
```

```python
path=r"../datas/AsiaZoo.txt"
```

```python
data=pd.read_csv(path,header=None)
```

```python
data
```

```python
data.columns=["国家","2006世界杯","2010世界杯","2007亚洲杯"]
```

```python
data
```

```python
kmn=KMeans(n_clusters=3)
x_train=data[["2006世界杯","2010世界杯","2007亚洲杯"]]
x_train
```

```python
y_new=kmn.fit(x_train).predict(x_train) #训练再预测
```

```python
y_new
```

```python
#绘制三维立体图，将各个国家画到3D图中
centers = kmn.cluster_centers_#KMeans算法得到的中心店
x,y,z = data['2006世界杯'],data['2010世界杯'],data['2007亚洲杯']

fig = plt.figure(figsize=(12,8)) #配置
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') #绘制3d

ax.scatter(x,y,z,c = kmn.predict(x_train),cmap='cool')  
ax.scatter(centers[:,0],centers[:,1],centers[:,2],c = kmn.predict(centers),
             cmap = "rainbow") 
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)
plt.show()
```
