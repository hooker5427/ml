### tensor 储存都是内存连续进行存储的， 但是可以改变数据的解析方式


```
x = tf.range(10)
print (x)
print( tf.reshape(x , [1,2,5,1]) ) # 改变视图 ，合法 
try:
  print (tf.reshape( x , [ 1,2,3]) ) # 不合法
except Exception  as e :
  print (e) 
  pass 
```

```
### 增加维度
x  = tf.range(10 ) 
x1 = tf.reshape( x ,(2,5) ) 
print (x1.shape )  


# axis =0处添加一维 
x2 = tf.expand_dims(x1 , 0 ) 
print (x2.shape )

# 最后进行添加一维 ，x1只有两位， 所有就是最后添加维度
x3 =  tf.expand_dims(x1 ,2 ) 
print (x3.shape )

x4 =  tf.expand_dims(x1 ,-1  ) 
print (x4.shape )
```

```
# 减少维度
# 均匀分布 
x  =  tf.random.uniform( [2,3] , maxval =10 , dtype = tf.int32  ) 
x 
```

```
# 高斯分布
x  =tf.random.normal(  shape= (10,10) , dtype= tf.float32 , mean= 0 ,stddev= 0.1 ) 
print (x.numpy())
```

```
x = tf.range( 10 )
x = tf.reshape ( x , (1, 2,5) ) 
print (x.numpy())
print (x.shape)


print ("----------------------------------")

#  减少维度
newx = tf.squeeze(x , axis = 0 ) 
print (newx.shape) 
print ( newx.numpy() ) 
```

广播

```
x = tf.random.normal((2,4) )
w = tf.random.normal( (4,2))
b  = tf.random.normal( [2] )

print (x.shape )
print (w.shape) 
print (b.shape )
print ((x@w).shape  )
print ('--------------------------')
print (x@w+b ) 
```
